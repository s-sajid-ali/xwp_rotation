```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.transform as skt

from xwp_rotation.rotate import pyvips_rotate
from xwp_rotation.rotate import rot_matrix
```

```python
size  = 256
angle = 45
```

```python
obj = np.zeros((size,size))
obj[size//2-size//4:size//2+size//4,
    size//2-size//4:size//2+size//4] = 1
```

```python
a = pyvips_rotate(obj,+angle)
b = skt.rotate(obj,+angle,order=3)
diff = np.abs(a - b)

plt.rcParams["figure.figsize"] = (20,5)
f,(ax1,ax2,ax3) = plt.subplots(1,3)
im1 = ax1.imshow(a, alpha = 0.5, origin='lower', cmap='coolwarm')
im2 = ax2.imshow(b, alpha = 0.5, origin='lower', cmap='coolwarm')
im3 = ax3.imshow(diff, alpha = 0.5, origin='lower', cmap='coolwarm')
f.colorbar(im1,ax=ax1)
f.colorbar(im2,ax=ax2)
f.colorbar(im3,ax=ax3)
```
